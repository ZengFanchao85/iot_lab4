```python
import pandas as pd
import seaborn as sns
import boto3
from io import StringIO
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/vehicle0.csv')
```

```python
fig2, ax2 = plt.subplots()
plt.plot(df['vehicle_x'], df['vehicle_y'], marker = "o", markerfacecolor = "green",label = 'position')
plt.xlabel("vehicle_x")
plt.ylabel('vehicle_y')
plt.legend()
```

```python
fig2, ax2 = plt.subplots()
plt.plot(df['timestep_time'], df['vehicle_speed'], marker = "o", markerfacecolor = "green",label = 'vehicle_speed')
plt.xlabel("timestep_time")
plt.ylabel('vehicle_speed')
plt.legend()
```

```python
fig2, ax2 = plt.subplots()
plt.plot(df['vehicle_noise'], df['vehicle_CO2'],marker = "o", markerfacecolor = "blue",label='CO2')
plt.xlabel("vehicle_noise")
plt.ylabel('vehicle_CO2')
plt.legend()
```

```python
fig2, ax2 = plt.subplots()
plt.plot(df['vehicle_noise'], df['vehicle_PMx'],marker = "o", markerfacecolor = "blue",label='PMX')
plt.xlabel("vehicle_noise")
plt.ylabel('vehicle_PMx')
plt.legend()
```

```python
fig2, ax2 = plt.subplots()
plt.plot(df['vehicle_speed'], df['vehicle_noise'], marker = "o", markerfacecolor = "green",label = 'noise')
plt.xlabel("vehicle_speed")
plt.ylabel('vehicle noise level')
plt.legend()
```
